```python
Q1: What is Estimation Statistics? Explain point estimate and interval estimate.
Answer-Estimation statistics is a branch of statistics concerned with estimating
population parameters based on sample data. It involves making inferences about 
population parameters, such as means, proportions, variances, and other characteristics,
using information obtained from a sample drawn from the population.
Point Estimate:
A point estimate is a single value used to estimate a population parameter.
It is based on sample data and is often calculated using a statistic, such 
as the sample mean or sample proportion, which is assumed to be a reasonable 
approximation of the population parameter

Interval Estimate:
An interval estimate provides a range of values within which the population parameter
is likely to lie, along with a level of confidence associated with that range. 
It is also known as a confidence interval. Interval estimates are more informative 
than point estimates because they give a sense of the variability and uncertainty 
associated with the estimation process

Q2. Write a Python function to estimate the population mean using a sample mean and standard
deviation.
Answer--import math

def estimate_population_mean(sample_mean, sample_std_dev, sample_size):
    # Z-score for 95% confidence level (assuming normal distribution)
    z_score = 1.96
    
    # Calculate the margin of error
    margin_of_error = z_score * (sample_std_dev / math.sqrt(sample_size))
    
    # Calculate the lower and upper bounds of the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    
    return (lower_bound, upper_bound)

# Example usage:
sample_mean = 50  # Sample mean
sample_std_dev = 10  # Sample standard deviation
sample_size = 100  # Sample size

lower_bound, upper_bound = estimate_population_mean(sample_mean, sample_std_dev, sample_size)
print("Estimated population mean lies between", lower_bound, "and", upper_bound, "with 95% confidence.")

Q3: What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing.
Answer--Hypothesis testing is a statistical method used to make inferences about a population parameter
based on sample data. It involves evaluating two competing hypotheses about the population parameter: 
    the null hypothesis (H0) and the alternative hypothesis (H1 or Ha). The process typically involves 
    assessing the evidence from the sample data to determine whether there is enough statistical evidence 
    to reject the null hypothesis in favor of the alternative hypothesis.

The null hypothesis (H0) typically represents the status quo or a default assumption about the population
parameter, while the alternative hypothesis (H1 or Ha) represents what the researcher is trying to find evidence for.

Hypothesis testing is used for several reasons:

Drawing conclusions about population parameters: Hypothesis testing allows researchers to draw 
conclusions about population parameters based on sample data. For example, it can help determine
whether there is a significant difference between the means of two populations or whether a new 
treatment is effective compared to a standard treatment.

Decision-making: Hypothesis testing provides a framework for making decisions based on evidence 
from sample data. By setting up hypotheses and conducting tests, researchers can make informed
decisions about accepting or rejecting hypotheses, which can have practical implications in 
various fields such as medicine, business, and social sciences.

Assessing the significance of results: Hypothesis testing helps assess the significance of
results obtained from sample data. It allows researchers to determine whether observed differences
or relationships are statistically significant or whether they could have occurred by random chance alone.

Q4. Create a hypothesis that states whether the average weight of male college students is greater than
the average weight of female college students.
Answer--The null hypothesis (H0) and the alternative hypothesis (H1) for comparing the average
weight of male college students and female college students can be formulated as follows:

Null Hypothesis (H0): The average weight of male college students is equal to or less than the 
average weight of female college students.
Alternative Hypothesis (H1): The average weight of male college students is greater than the
average weight of female college students.

Mathematically, we can represent these hypotheses as:

H0: μ_male <= μ_female
H1: μ_male > μ_female

Here, μ_male represents the population mean weight of male college students, 
and μ_female represents the population mean weight of female college students.

Q5. Write a Python script to conduct a hypothesis test on the difference between two population means,
given a sample from each population.
Answer--from scipy import stats

def conduct_hypothesis_test(sample1, sample2, alpha=0.05):
    # Perform independent two-sample t-test
    t_statistic, p_value = stats.ttest_ind(sample1, sample2, equal_var=True)

    print("Independent Two-Sample T-Test Results:")
    print("t-statistic:", t_statistic)
    print("p-value:", p_value)

    # Check if the p-value is less than the significance level
    if p_value < alpha:
        print("Reject the null hypothesis. There is sufficient evidence to conclude that the population means are different.")
    else:
        print("Fail to reject the null hypothesis. There is not sufficient evidence to conclude that the population means are different.")

# Example usage:
sample1 = [72, 68, 74, 71, 69]  # Sample from population 1
sample2 = [65, 67, 63, 64, 66]  # Sample from population 2

conduct_hypothesis_test(sample1, sample2)

Q6: What is a null and alternative hypothesis? Give some examples.
Answer--In statistical hypothesis testing, the null hypothesis (denoted as H0) and the 
alternative hypothesis (denoted as H1 or Ha) are two competing statements regarding a population 
parameter or a statistical relationship between variables. These hypotheses help guide the hypothesis 
testing process and the interpretation of results.

Null Hypothesis (H0):

The null hypothesis represents the default assumption, often stating that there is no effect, 
no difference, or no relationship in the population.
It is the hypothesis that the researcher aims to test against.
The null hypothesis typically states that any observed differences or effects in the sample data
are due to random variation or chance.
The null hypothesis is usually denoted as H0.
Example:

H0: The mean score of students who took the new teaching method is equal to the mean score of 
students who took the traditional teaching method.
Alternative Hypothesis (H1 or Ha):

The alternative hypothesis contradicts the null hypothesis.
It represents what the researcher is trying to find evidence for, often asserting the presence 
of an effect, difference, or relationship in the population.
The alternative hypothesis may state that observed differences in the sample data are not due 
to random variation and are instead meaningful.
The alternative hypothesis can take different forms depending on the research question, such as one-tailed or two-tailed.
Examples:

H1: The mean score of students who took the new teaching method is greater than the mean score\
of students who took the traditional teaching method.
H1: There is a significant difference in the mean heights of men and women in the population.

Q7: Write down the steps involved in hypothesis testing.
Answer--Hypothesis testing is a statistical method used to make inferences about 
population parameters based on sample data. The general steps involved in hypothesis testing are as follows:

Formulate the Null Hypothesis (H0) and Alternative Hypothesis (H1):

Define the null hypothesis, which typically represents the status quo or the absence 
of an effect, difference, or relationship.
Formulate the alternative hypothesis, which contradicts the null hypothesis and represents
what the researcher is trying to find evidence for.
Choose the Significance Level (α):

Select the significance level, denoted as α, which represents the probability of rejecting the 
null hypothesis when it is actually true. Commonly used significance levels include 0.05, 0.01, and 0.10.
Select the Appropriate Test Statistic:

Choose a statistical test that is appropriate for the hypothesis being tested, the type of data
collected, and the research question. Examples of statistical tests include t-tests, chi-square 
tests, ANOVA, correlation tests, and regression analysis.
Collect Sample Data:

Collect relevant data from a representative sample or samples, ensuring that the data collection
process is unbiased and that the sample size is appropriate for the chosen statistical test.
Calculate the Test Statistic:

Compute the test statistic using the sample data and the chosen statistical test. 
The test statistic quantifies the difference or relationship observed in the sample
data and provides a basis for comparing it to the null hypothesis.
Determine the Critical Region or Critical Value:

Determine the critical region or critical value based on the chosen significance
level and the distribution of the test statistic under the null hypothesis.
The critical region represents the range of values for the test statistic that would lead to rejection of the null hypothesis.
Make a Decision:

Compare the calculated test statistic to the critical region or critical value.
If the test statistic falls within the critical region, reject the null hypothesis
in favor of the alternative hypothesis.
If the test statistic does not fall within the critical region, fail to reject the null hypothesis.
Draw Conclusions:

Based on the decision made in step 7, draw conclusions regarding the null hypothesis and the research question.
Interpret the results in the context of the study and consider the implications for future research or decision-making.
Report the Results:

Report the findings of the hypothesis test, including the calculated test statistic, the decision
made regarding the null hypothesis, and any relevant measures of effect size or uncertainty.
Q8. Define p-value and explain its significance in hypothesis testing.
Answer--The p-value, or probability value, is a crucial concept in hypothesis testing
that quantifies the strength of the evidence against the null hypothesis (H0) and helps
determine whether to reject or fail to reject the null hypothesis based on sample data.

Here's a definition and explanation of the p-value and its significance in hypothesis testing:

Definition:
The p-value is the probability of obtaining test results as extreme as or more extreme 
than the observed results, under the assumption that the null hypothesis is true. 
In other words, it measures the likelihood of observing the data or something more extreme if the null hypothesis is correct.

Significance in Hypothesis Testing:

A small p-value indicates that the observed data are unlikely to occur if the null hypothesis
is true, suggesting strong evidence against the null hypothesis.
Conversely, a large p-value suggests that the observed data are likely to occur even if the null
hypothesis is true, indicating weak evidence against the null hypothesis.
Interpretation:

If the p-value is less than or equal to the chosen significance level (α), typically 0.05, 
the results are considered statistically significant.
A small p-value (less than α) leads to the rejection of the null hypothesis in favor of the
alternative hypothesis, suggesting that the observed results are unlikely to occur due to random chance alone.
A large p-value (greater than α) leads to the failure to reject the null hypothesis, indicating
that the observed results are consistent with what would be expected under the null hypothesis.
Significance Level (α):

The significance level (α) is the threshold used to determine statistical significance.
It represents the probability of rejecting the null hypothesis when it is actually true.
Commonly used significance levels include 0.05, 0.01, and 0.10, although the choice depends
on the specific requirements of the study and the field of research.
Q9. Generate a Student's t-distribution plot using Python's matplotlib library, with the degrees of freedom
parameter set to 10.
Answer--import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Define degrees of freedom
df = 10

# Generate x values
x = np.linspace(-4, 4, 1000)

# Calculate the probability density function (PDF) for the Student's t-distribution
pdf = t.pdf(x, df)

# Plot the PDF
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label=f'Student\'s t-distribution (df={df})', color='blue')
plt.title('Student\'s t-distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two
random samples of equal size and a null hypothesis that the population 
means are equal.
Answer--import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Define degrees of freedom
df = 10

# Generate x values
x = np.linspace(-4, 4, 1000)

# Calculate the probability density function (PDF) for the Student's t-distribution
pdf = t.pdf(x, df)

# Plot the PDF
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label=f'Student\'s t-distribution (df={df})', color='blue')
plt.title('Student\'s t-distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
Q11: What is Student’s t distribution? When to use the t-Distribution.
Answer--import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Define degrees of freedom
df = 10

# Generate x values
x = np.linspace(-4, 4, 1000)

# Calculate the probability density function (PDF) for the Student's t-distribution
pdf = t.pdf(x, df)

# Plot the PDF
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label=f'Student\'s t-distribution (df={df})', color='blue')
plt.title('Student\'s t-distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
Q12: What is t-statistic? State the formula for t-statistic.
Answer--
```
